```
import numpy
import pandas as pd
import os
import io
from google.colab import files
import matplotlib.pyplot as plt
```

## D_Error


```
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['D_error.csv']))
```

```

df.head()
```

```
plt.ylim(-0.515875, -0.515845)
plt.plot(df['Step'], df['Value'])
```

```
for i in range(len(df['Step'])):
  fig = plt.figure(figsize = (10, 10))
  plt.ylim(-0.515875, -0.515845)
  plt.plot(df['Step'][:i], df['Value'][:i])
  os.chdir('/content')
  save_dir = os.path.join("graphs/", '{}/'.format("D_error"))
  if not os.path.exists(save_dir):
    os.makedirs(save_dir)
  fig.savefig('{}/{:04d}.png'.format(save_dir, i))
  plt.close(fig)
```
